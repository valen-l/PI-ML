```python
#Importamos las librerias para nuestro modelo de recomendacion
from sklearn.preprocessing import OrdinalEncoder
from sklearn.neighbors import NearestNeighbors
```

```python
# Preprocesamos de datos
movies['popularity'] = pd.to_numeric(movies['popularity'], errors='coerce')
movies['popularity'] = movies['popularity'].fillna(0)  # Rellenar valores faltantes en la columna 'popularity'
```

```python
# Codificar la columna 'popularity' de manera ordinal
encoder = OrdinalEncoder()
movies['popularity_encoded'] = encoder.fit_transform(movies['popularity'].values.reshape(-1, 1))
```

```python
# Entrenar el modelo KNN
k = 6  # Número de vecinos más cercanos a considerar (excluyendo la película de entrada)
knn_model = NearestNeighbors(n_neighbors=k, metric='euclidean')
knn_model.fit(movies[['popularity_encoded']])
```

```python
# Función para obtener las recomendaciones de películas
def get_recommendations(movie_title):
    # Obtener el índice de la película con el título dado
    idx = movies[movies['title'] == movie_title].index[0]

    # Obtener las distancias y los índices de los vecinos más cercanos
    distances, indices = knn_model.kneighbors(movies.loc[[idx], ['popularity_encoded']])

    # Obtener los índices de las películas más similares (excluyendo la película de entrada)
    top_indices = indices.flatten()[1:k+1]

    # Obtener las películas recomendadas ordenadas por popularidad descendente
    recommendations = movies.iloc[top_indices].sort_values('popularity', ascending=False)[['title', 'overview']]

    return recommendations

```

```python
#Revisamos que sea correcto su funcionamiento
get_recommendations('Charlie and the Chocolate Factory')
```
